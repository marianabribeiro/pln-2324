```python
from gensim.utils import tokenize
import re
from gensim.models import Word2Vec
from gensim.utils import tokenize
from gensim.models import KeyedVectors
```

##### Abertura e Leitura dos Ficheiros

```python
f1=open("drive/MyDrive/Harry_Potter_Camara_Secreta-br.txt", encoding='utf-8')
f2= open("drive/MyDrive/Harry_Potter_e_A_Pedra_Filosofal.txt", encoding='utf-8')

text1=f1.read()
text2=f2.read()
```

##### Preparação dos textos para o modelo

```python
linhas1=text1.split("\n")
linhas2=text2.split("\n")
tokens=[]


for linha in linhas1:
    #linha=re.sub(r"[!,-?–\.]+", "", linha) #podiamos fazer linha.lower
    linha=list(tokenize(linha, lower=True))
    tokens.append(linha) # antes : linha.split()
for linha in linhas2:
    linha=list(tokenize(linha, lower=True))
    tokens.append(linha)
```

##### Modelo

```python
# Foi testado vector_size=300; epochs=20
model=Word2Vec(tokens, vector_size=300, window=5, min_count=1, sg=1, epochs=20, workers=3)
model.wv.most_similar("harry")
```

```python
# Foi testado vector_size=250; epochs=20
model=Word2Vec(tokens, vector_size=250, window=5, min_count=1, sg=1, epochs=20, workers=3)
model.wv.most_similar("harry")
```

```python
# Foi testado vector_size=100; epochs=20
model=Word2Vec(tokens, vector_size=100, window=5, min_count=1, sg=1, epochs=20, workers=3)
model.wv.most_similar("harry")
```

```python
# Foi testado vector_size=100; epochs=15
model=Word2Vec(tokens, vector_size=100, window=5, min_count=1, sg=1, epochs=15, workers=3)
model.wv.most_similar("harry")
```

```python
# Foi testado vector_size=100; epochs=5
model=Word2Vec(tokens, vector_size=100, window=5, min_count=1, sg=1, epochs=5, workers=3)
# O modelo inicial vector_size=100; epochs=5 foi o que pareceu ligar melhor a palavra harry,
#uma vez que o relaciona com outras personagens e nao com palavras "random"
model.wv.most_similar("harry")
```
